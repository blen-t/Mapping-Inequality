```python
#Lina Abdo
#year_low
#change all years to 4 digits
#remove apostrophes
#remove spaces between dashes
#replace empty cells with N/A

for year in df["year_low"]:
    year = str(year)
    if len(year) == 2:
        print("19" + year)
    elif len(year) == 4:
        print(year)
    elif (" - ") in year:
        if len(year) == 9:
            year = year.replace(" - ", "-19")
            print(year)
        else:
            year = year.replace(" - '", "-19")
            print("19" + year)
    elif ("-") in year:
        if len(year) == 5:
            year = year.replace("-", "-19")
            print("19" + year)
        elif len(year) == 6:
            year = year.replace("-'", "-19")
            print("19" + year)
        elif len(year) == 7:
            year = year.replace("-", "-19")
            print(year)
        else:
            print(year)
    else:
        print("N/A")
```

```python
#Lina Abdo
#SV_range_low
#remove commas
#replace to with -
#remove extra symbols (%, '', .)
#add $ before numbers
#replace empty cells with N/A

for number in df["SV_range_low"]:
    number = str(number)
    number = number.replace(" to ", "-").replace(" - ", "-").replace("- ", "-")
    number = number.replace(",", "").strip("''").strip("%")
    if number.startswith("11000"):
        print("$11000")
    elif (".") in number:
        print("$9000")
    elif ("&") in number:
        print("$" + number)
    elif len(number) == 4:
        print("$" + number)
    elif ("-") in number:
        if ("-$") in number:
            print(number)
        else:
            number = number.replace("-", "-$")
            print("$" + number)
    else:
        print("N/A")
```

```python
#Dua Kheiry
#Date
#changes each date into a string
#replaces all the single quotation marks into dashes
#replaces each month with it's numerical equivalent
#changes different formatting of date
#replaces # and empty cells into N/A

for date in df["Date"]:
    date = str(date)
    if "'" in date: #this statement doesn't work
        date = date.replace("'", "-")
        print(date)
    elif date.startswith("October"):
        date = date.replace("October 15th 1937", "10-15-1937")
        print(date)
    elif date.startswith("June"):
        date = date.replace("June 12 1937", "06-12-1937")
        print(date)
    elif "May" in date:
        date = date.replace ("3-May-37", "05-03-1937")
        date = date.replace ("4-May-37", "05-04-1937")
        date = date.replace ("5-May-37", "05-05-1937")
        print(date)
    elif date.startswith("August"):
        date = date.replace ("August 13, 1937", "08-13-1937")
        print(date)
    elif "#" in date:
        print("N/A")
    elif "July 1937" in date:
        date = date.replace("July 1937", "07/1937")
        print(date)
    elif "37" in date:
        date = date.replace("37", "1937")
        print(date)
    elif "nan" in date:
        print("N/A") 
    else: 
        print(date)
```
